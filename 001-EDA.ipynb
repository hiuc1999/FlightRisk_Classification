## Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
emp_df = pd.read_csv('data/Human_Resources.csv')
emp_df.head()
```

```python
# Drop columns in which all rows have the same values
# Also drop EmployeeNumber which does not contribution to the problem
emp_df.drop(['EmployeeCount', 'StandardHours','Over18','EmployeeNumber'], axis=1,inplace=True)
```

```python
# Encode binary variavble (Attrition, OverTime)
for i in ['Attrition', 'OverTime']:
    emp_df[i] = np.where(emp_df[i] == 'Yes',1,0)
```

```python
# Compare the employees who stayed and left
left_df = emp_df[emp_df['Attrition'] == 1]
stayed_df = emp_df[emp_df['Attrition'] == 0]

print('Total:' , len(emp_df))
print(f'Number of employees who left: {len(left_df)} ({round(100*len(left_df)/len(emp_df),2)}%)')
print(f'Number of employees who left: {len(stayed_df)} ({round(100*len(stayed_df)/len(emp_df),2)}%)')
```

```python
emp_df.select_dtypes(include=np.number).columns.to_list()
```

```python
con_var = ['Age',
 'DailyRate',
 'DistanceFromHome',
 'Education',
 'EnvironmentSatisfaction',
 'HourlyRate',
 'JobSatisfaction',
 'MonthlyIncome',
 'MonthlyRate',
 'NumCompaniesWorked',
 'PercentSalaryHike',
 'PerformanceRating',
 'StockOptionLevel',
 'TotalWorkingYears',
 'TrainingTimesLastYear',
 'WorkLifeBalance',
 'YearsAtCompany',
 'YearsInCurrentRole',
 'YearsSinceLastPromotion',
 'YearsWithCurrManager'] # continiuous variables

cat_var = [i for i in emp_df.columns if i not in con_var] # categorical variables
```

```python
fig, axes = plt.subplots(4,5,figsize=(20,20))

for i in range(len(con_var)):
    plt.subplot(5,5,i+1)
    sns.boxplot(data = emp_df, y = con_var[i], x = 'Attrition' )
    plt.xlabel(None)
    plt.ylabel(None)
    plt.title(con_var[i])
```

From the above boxplot, comparing distribution of continuous featurs between employees who stayed and employees who left.
Generally speaking,
- Age:  Employees who stayed are older than who left.
- DailyRate:  Employees who stayed have higher daily rate.
- DistanceFromHome: Employees who stayed live closer to the company.
- EnvironmentalSatisfaction & JobSatisfaction: Employees who stayed are more satisfied with their job.
- MonthlyIncome: Employee who stayed have higher monthly income.
- TotalWorkingYears & YearsAtCompany: Employee who stayed have been in the workforce and working in this company for longer time. 
- YearsInCurrentRole & YearsWithCurrManager: Employee who stayed has worked for longer time in current position and= with current manager.

```python
corr = emp_df.corr()
f, ax = plt.subplots(figsize=(20,20))
sns.heatmap(corr, annot=True)
```

```python
fig, axes = plt.subplots(6,2,figsize=(20,20), constrained_layout = True)

for i in range(len(cat_var)):
    plt.subplot(6,2,i+1)
    sns.countplot(data=emp_df,x=cat_var[i], hue = 'Attrition' )
    plt.xlabel(None)
    plt.ylabel(None)
    plt.xticks(rotation=10)
    plt.title(cat_var[i])
    

plt.show()
```

- Single employees are more likely to leave than married/ divorced employees
- Sales Representatives tend to leave comparing with other roles.
- Employees who work overtime are more likely to leave than who do not need to work overtime. 
- JobLevel: Employee who stayed have higher job levels.

```python
fig, axes = plt.subplots(2,1,figsize=(14,14))
plt.subplot(2,1,1)
sns.kdeplot(left_df['DistanceFromHome'], label = 'Employees who left', shade = True, color = 'r')
sns.kdeplot(stayed_df['DistanceFromHome'], label = 'Employees who stayed', shade = True, color = 'b')
plt.xlabel('Distance From Home')

plt.subplot(2,1,2)
sns.kdeplot(left_df['YearsWithCurrManager'], label = 'Employees who left', shade = True, color = 'r')
sns.kdeplot(stayed_df['YearsWithCurrManager'], label = 'Employees who stayed', shade = True, color = 'b')

plt.xlabel('Years With Current Manager')

plt.show()
```
